# 2.一下子可以画两个图

```python
x1 = np.linspace(0.0,5.0) #默认是50个点
```

```python
x2 = np.linspace(0.0,2.0) #0到2中间按相同间隔取50个点
```

```python
print(x1)
```

```python
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)
```

```python
plt.subplot(2,1,1)  # 2行1列 第一个图
plt.plot(x1,y1,'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2,1,2) # 2行1列 第二个图
plt.plot(x2,y2,'.-g')
plt.xlabel('time(s)')
plt.ylabel('Undamped')
```

# 9. Error bar charts
## plotting a single data point (the mean value of dataset),and an error bar to represent the overall distribution of data
## this helps us to get a general idea to how accurate a measurement is 
## errorbar() function
## keywords:
## fmt : 这是画线的格式，如果None,那么error bars are plotted with no line connecting them
## ecolor
## elinewidth:描述了error bars的线宽
## capsize: this specifies the size of the cap of the error bars in pixels

```python
x = np.arange(0,4,0.2)
y = np.exp(-x)
e1 = 0.1 * np.abs(np.random.randn(len(y)))
plt.errorbar(x,y,yerr = e1,fmt = '.-')
```

```python
e2 = 0.1 * np.abs(np.random.randn(len(y)))
plt.errorbar(x,y,yerr = e1,xerr = e2,fmt ='.-',capsize = 0)
```

#### this is symmetrical error bar

```python
plt.errorbar(x,y,yerr =[e1,e2],fmt='--')
```

#### this is asymmetrical error bar

# 11.饼图呀大饼图
## pie() plot pie charts from an array x
## keywords：
## explode:it is an array of the same length as that of x,each of its values specify th radius fraction with which to offest the wedge from the center of the pie
## colors
## labels,labeldistance:this is a list of labels,one for each of the X values,labeldistance is the radial distance at which the labels are drawn
## autopct,pctdistance:this formatting string or function is used o label wedges with their numeric values. pctdistance is the ratio between the center of the pie and the start of the text
## shadow:draws a shadow for weges

```python
plt.figure(figsize =(3,3))
x =[45,35,20]
labels =['cats','dogs','fishes']
plt.pie(x,labels = labels)
```

```python
plt.figure(figsize =(4,4))
x =[4,9,21,55,30,18]
labels = ['Swiss','Austria','Spain','Italy','France','Benelux']
explode =[0.2,0.1,0,0,0.1,0]
plt.pie(x,labels=labels,explode = explode,autopct ='%1.1f%%')
```

# 13. Polar Charts
## polar() plots polar charts
## its parameters are two lists
## theta for the angular coordinates and r for the radial coordinates

```python
theta = np.arange(0,2,1/180)*np.pi # theta is an array of 360 values,from 0 to 2pi
```

```python
plt.polar(3*theta,theta/5)   # 螺旋
plt.polar(theta,np.cos(4*theta)) # draw a polar rose
plt.polar(theta,[1.4]*len(theta)) # draw a circular line
```

## navigation toolbar with polar plots
## rgrid() and thetagrid() functions to control the radial and angular grids
## 1.rgrid() arguments we can pass
## radii :the radial distances at which the grid lines should be drawn
## labels:the labels to display at radii grid
## angle:the angle at which the labels are displayed
## 2.thetagrids() arguments we can pass
## angles:the location where to draw the labels
## labels:specifies the labels to be printed at given angles
## frac:the polar axe radius fraction at which we want the label to be drawn

```python
theta = np.arange(0,2,1/180)*np.pi
```

```python
r = np.abs(np.sin(5*theta)-2*np.cos(theta))
```

```python
plt.polar(theta,r)
plt.thetagrids(range(45,360,90))
plt.rgrids(np.arange(0.2,3.1,0.7),angle = 0)
```

# 14.如何在图中添加文本
## 1.text()  write a string at an arbitrary position 
##  plt.text(x,y,text)
## 2.annotate() provides functionality to make annotation easy
## plt.annotate(text,xy,xytext,arrowprops,shrink)
## xy is the ponit we want to annotate,xytext is the position where we want to place the annotation,both expressed in (x,y) format
## arrowprops specify the arrow properties
## shrink is used to make some space from the points and the arrow tip and base
## 3.箭头画法 arrow(x,y,dx,dy)
## (x,y) is the arrow origin
## (dx,dy) the distance at which the head is to be placed
## arrow starts at (x,y) and ends at (x+dx,y+dy)
## this function is quite hard to use,we can use annotate() without text to only create an arrow

```python
x = np.arange(0,2*np.pi,0.01)
y = np.sin(x)
plt.plot(x,y)
plt.text(0.1,-0.04,'sin(0) = 0')
```

## an example of the usage of annotate()

```python
y = [13,11,13,12,13,10,30,12,11,13,12,12,12,11,12]
```

```python
plt.plot(y)
plt.ylim(ymax = 35)
plt.annotate('this spot must really\nmean something',xy = (6,30),xytext =(8,31.5),arrowprops = dict(facecolor ='green',shrink =0.05))
```

## an example of arrow drawing

```python
plt.axis([0,10,0,20])
arrstyles =['-','->','-[','<-','<->','fancy','simple','wedge']
for i ,style in enumerate(arrstyles):
    plt.annotate(style,xytext =(1,2+2*i),xy=(4,1+2*i),arrowprops = dict(arrowstyle = style))
connstyles =["arc","arc,angleA=10,armA=30,rad =15","arc3,rad =0.2","arc3,rad=-0.2","angle","angle3"]
for i ,style in enumerate(connstyles):
    plt.annotate("",xytext=(6,2+2*i),xy = (8,1+2*i),arrowprops = dict(arrowstyle ='->',connectionstyle = style))
```
